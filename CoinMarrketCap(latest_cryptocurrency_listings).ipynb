```python
import requests
import pandas as pd
```

```python
API_KEY = "4c48da95-55a9-4af2-9ecd-0597fa01fb2a"

# Base URL for CoinMarketCap API
BASE_URL = "https://pro-api.coinmarketcap.com"

# endpoint to get latest cryptocurrency listings
# Reference: https://coinmarketcap.com/api/documentation/v1/#operation/getV1CryptocurrencyCategory
ENDPOINT = "/v1/cryptocurrency/listings/latest"

# Headers with your API key
headers = {
  "X-CMC_PRO_API_KEY": API_KEY
}

# Send the request and get the response
response = requests.get(url=f"{BASE_URL}{ENDPOINT}", headers=headers)

# Check for successful response
if response.status_code == 200:
  # Parse the JSON response
  data = response.json()
  # Access the data you need
  print(data)
else:
  print(f"Error: {response.status_code}")
```

```python
#Extract data from response and then transfrom to pandas dataframe
Data = data['data']
timestamp = data['status']['timestamp']
df_data = []
for item in Data:
    id = item['id']
    name = item['name']
    symbol = item['symbol']
    max_supply = item['max_supply']
    price = item['quote']['USD']['price']

    item_data = {
      'id' : id,
      'timestamp' : timestamp,
      'name' : name,
      'symbol' : symbol,
      'max_supply' : max_supply,
      'price' : price
    }

    df_data.append(item_data)
df = pd.DataFrame(df_data)
print(df)
df.to_csv('df.csv', index = False)
```
